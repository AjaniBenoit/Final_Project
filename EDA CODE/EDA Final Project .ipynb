```python
# Files to load 
STSI_load = "Resources/STSI_Combined .csv"

# Read the CSV into a Dataframe 
STSI_df = pd.read_csv(STSI_load)
STSI_df
```

```python
updated_merged_df=merged_df.drop("Fatalities", axis='columns')
updated_merged_df
```

```python
updated_merged_df=updated_merged_df.drop("Alcohol-Impaired Driving Fatalities (BAC", axis='columns')
updated_merged_df
```
